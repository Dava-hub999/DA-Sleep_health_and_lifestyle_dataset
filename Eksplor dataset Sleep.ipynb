```python
import matplotlib.pyplot as plt

# Membuat barplot
plt.figure(figsize=(8, 6))
plt.bar(df['Gender'], df['Stress Level'])

# Menambahkan judul dan label sumbu
plt.title('Contoh Barplot')
plt.xlabel('Kategori')
plt.ylabel('Jumlah')

# Menampilkan plot
plt.show()
```

```python
# Pemilik stress level paling tinggi
StressLevel = df.groupby(['Gender', 'Stress Level']).size().unstack(fill_value=0)

plt.figure(figsize=(8, 6))
StressLevel.plot(kind='barh', color=['skyblue', 'pink', 'orange'])

plt.title('Stress Level by Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.legend(title='Stress Level', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
#
bmi_counts = df['BMI Category'].value_counts()

plt.figure(figsize=(8, 8))
plt.pie(bmi_counts, labels=bmi_counts.index, autopct='%1.1f%%', startangle=140)
plt.title("BMI Category Distribution")
plt.show()
```

```python
import numpy as np

gender_stress = df.groupby('Gender')['Stress Level'].mean()
x = np.arange(len(gender_stress))
plt.figure(figsize=(8, 6))
plt.bar(x, gender_stress, width = 0.35, label='Stress Level', color=['skyblue', 'green'])
plt.xlabel('Gender')
plt.ylabel('Average Stress Level')
plt.title('Stress Level by Gender')
plt.xticks(x, gender_stress.index)
plt.legend(title='Stress Level', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```
